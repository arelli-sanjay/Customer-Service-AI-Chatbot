### Import Libraries

```python
import numpy as np
import pandas as pd
```

### NLP

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
ps = WordNetLemmatizer()
```

```python
stopwords = stopwords.words('english')
```

```python
def clean_row(row):

    row = row.lower()
    row = re.sub('[^a-zA-Z]', '', row)
    
    token = row.split()
    
    Issue = [ps.lemmatize(word) for word in token if not word in stopwords]
    
    cleaned_Issue = ' '.join(Issue)
    
    return cleaned_Issue
```

```python
dataset['user_query']
```

```python
dataset['user_query'] = dataset['user_query'].apply(lambda x : clean_row(x))
```

```python
dataset['user_query']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=3000, lowercase=False, ngram_range=(1,2))
```

```python
x = dataset['user_query']
y = dataset['intent']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
vec_x_train = vectorizer.fit_transform(x_train)
```

```python
vec_x_train = vec_x_train.toarray()
```

```python
type(vec_x_train)
```

```python
vec_x_test = vectorizer.transform(x_test)
```

```python
vec_x_test = vec_x_test.toarray()
```

```python
vec_x_train.shape,vec_x_test.shape
```

```python
vec_x_train
```

```python
vec_x_test
```

```python
training_data = pd.DataFrame(vec_x_train,columns = vectorizer.get_feature_names_out())
testing_data = pd.DataFrame(vec_x_test,columns = vectorizer.get_feature_names_out())
```

```python
training_data
```

```python
testing_data
```

### Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
```

```python
clf.fit(vec_x_train, y_train)
```

```python
import warnings
warnings.filterwarnings("ignore", message="X has feature names")
```

```python
y_pred = clf.predict(testing_data)
```

### Confusion Matrix and Accuracy

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

### Sample Example of Testing Data

```python
User_Query = input("Enter query: ")

response = clean_row(str(User_Query))

pred = clf.predict(vectorizer.transform([response]).toarray())[0]

if pred:
    reply = dataset[dataset['intent'] == pred]['response'].sample(1).values[0]
    print(f"\nPredicted Intent: {pred}")
    print(f"Chatbot Response: {reply}")
else:
    print("Not Found Any Response")
```
